```python
import pandas as pd

data1 = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data1)

print(df)
print(df.head(2))
print(df.tail(2))
print(df.info())
print(df.describe())
```

```python
import pandas as pd

data1 = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data1)

print(df)
```

```python
#loc 1 -> index of row, second -> cols name
print(df.loc[1,"Name"])

#iloc row index -> 0 and col index -> 0
print(df.iloc[1,1])

#get rows
print(df.loc[[0,1,2]])

#get rows
print(df.iloc[[1,2,3]])

#muti rows and cols
print(df.iloc[[0,1],[0,1]])

#slicing
print(df.iloc[0:3])
```

```python
#filter one greater
print(df[df['Emp_ID']>104])
#filter equal
print(df[df['Emp_ID']==106])
#salary between 30000 & 60000
print(df[(df["Salary"]>30000 )&( df["Salary"]<60000)])
```

```python
#sorting ascending
print(df.sort_values("Salary",ascending=True))

#sorting descending
print(df.sort_values("Salary",ascending=False))
```

```python
#two col
print(df.sort_values(["Salary","Experience"]))
```
