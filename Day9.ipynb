```python
#numpy | list
import numpy as np
import time

start= time.time()
sum1=0
for i in range(100000):
    sum1+= i*2
end = time.time()
#for list
print("for list : ",sum1,end-start)

#for numpy | array(array works on every element)
start1=time.time()
sum2=0
arr = np.arange(100000)*2
sum2=arr.sum()
end1=time.time()
print("for array : ",sum2,end1-start1)
```
