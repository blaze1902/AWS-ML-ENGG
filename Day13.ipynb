```python
import pandas as pd
```

```python
l=[10,20,30,40,50]  #list
df=pd.Series(l)  #use list i pd
print(df)    


#example data
df=pd.Series(data=l)
print(df)

#example index
df=pd.Series(data=l , index=['aditya','atul','sameer','rudra','prince'])
print(df)

#example key=index | value=value
d={'aditya':13,'atul':63,'sameer':12,'rudra':15,'prince':23}
df=pd.Series(data=d)
print(df)
```

```python
#2D
data={
    "name":['aditya','atul','sameer','rudra','prince'],
    "age":[10,20,30,40,50]
}

#1d -> represent
df=pd.Series(data)
print(df)

#2d -> represent
df=pd.DataFrame(data)  #we use DataFrame() to convert into 2D
print(df)
```

```python
data=[['aditya',13],['atul',63],['sameer',23],['rudra',26],['prince',56]]
df=pd.DataFrame(data)
print(df)

# add columns name
df=pd.DataFrame(data,columns=['name','age'])
print(df)

#add indexing name
df=pd.DataFrame(data,index=['a','b','c','d','e'])
print(df)
```
