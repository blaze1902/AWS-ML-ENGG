LIST METHOD

```python
list=[10,20,30]
print(list)

#append ---> add element at last
#append(value)
list.append(40)
print("After append:", list)

#insert ---> add element at specific index
#insert(index, value)
list.insert(1, 15)
print("After insert: ",list)

#length of list --> len(list)
print("Length of list: ", len(list))
```

```python
#list element access
print(list[0]) #first element
print(list[1]) #second element
print(list[-1]) #last element
```

FUNCTION

A block of code that use to do specific task

```python
def Hello():
    print("Hello function")

Hello()
Hello()
```

```python
def name(a,b):
    print(a,b)
name("Aditya","Rao")
```

```python
def name(*a):
    print(a)
name("Aditya","Rao","Atul")
name("Aditya","Rao")
```

```python
def name(*a):
    for i in a:
        print(i)
name("Aditya","Atul")
name("A","B","C","D")
```
