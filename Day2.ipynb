1 ---> ARITHMATIC OPERATOR

```python


num1=10
num2=20
print(num1+num2)  #30

```

```python
#example 2

num1=200
num2=20
print(num1-num2)  #180
```

```python
#example 3

num1=5
num2=3
print(num1*num2)  #15
```

```python
#example 4

num1=100
num2=10
print(num1/num2)  #10.0
```

```python
#example 5  

num1=10
num2=3
print(num1%num2)  #1
```

```python
#example 6
num1=20
num2=3
print(num1//num2)  #floor value 
print(num1/num2)   #normal division
print(num1**num2)  #power
print(num1%num2)   #modulus
```

2 ---> RELATIONAL OPERATOR ( <,>,>=,<=,==,!=)


```python
num1=10
num2=20
print(num1>num2) #geater than
print(num1<num2) #less than
print(num1>=num2) #greater than or equal to
print(num1<=num2) #less than or equal to
print(num1==num2) #equal to
print(num1!=num2) #not equal to
```

3 ---> ASSIGNMENT OPERATOR

```python
num1=10
num2=3
num1 += num2  #num1=num1+num2
print(num1) 
```

```python
num1=10
num2=3
num1 -= num2  #num1=num1-num2
print(num1)
num1 *= num2 #num1=num1*num2
print(num1)
num1 /= num2  
print(num1)
num1%=num2  
print(num1)
```
