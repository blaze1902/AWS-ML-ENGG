```python
#function calculator --> add,sub
def cal(a,b,operator):
    if operator == "+":
        print(a+b)
    elif operator == "-":
        print(a-b)
    elif operator == "*":
        print(a*b)
    elif operator == "/":
        print (a/b)
    elif operator == "%":
        print(a%b)
    else:
        print(a//b)

cal(10,20,"+")
cal(10,20,"-")
cal(10,20,"/")
cal(10,20,"%")
```

```python
#using eval function 
#eval("10 + 20")


def cal(*a):
    print(eval(f"{a[0]} {a[-1]} {a[1]}"))

cal(10,20,"+")
cal(10,20,"-")
cal(10,20,"/")
cal(10,20,"%")
```
