```python
class Student:
    def __init__(self,*args):
        self.data=args

s=Student({"users":["Aditya","Blaze","Golu","Atul"]},{"address":{"home-town":"kukas","office":"Jaipur"}})
print(s.data)
print(s.data[0])
print(s.data[1])
print(s.data[0]["users"])
print(s.data[1]["address"])


#loops under access of dict
for i in s.data[0]["users"]:
    print(i)

for i in s.data[1]["address"].keys():
    print(i)
    print(s.data[1]["address"][i])
    
```

```python
class Student:
    def __init__(self,name,roll_no):
        self.name=name
        self.roll_no=roll_no

    #get
    def getName(self):
        print(self.name)
    #update
    def updateName(self):
        self.name="hello"
        print(self.name)
    #delete
    def deleteName(self):
        del self.name
        print(self.name)
    #create
    def createName(self):
        self.full_name = "Aditya Rao"
        print(self.full_name)

s=Student("nikhil",123)
s.getName()
s.createName()
s.updateName()
```
