Inheritance

```python
class User1:
    def __init__(self,pen,pencil):
        print("Constructor of User1")
        self.pen=pen
        self.pencil=pencil
    def penGet(self):
        print("pen is working :",self.pen)
    def pencilGet(self):
        print("pencil is working :",self.pencil)

class User2(User1):
    def __init__(self,pen,pencil,erasor):
        print("Constructer of User2")
        self.pen=pen
        self.pencil=pencil
        self.erasor=erasor

#object
u=User2("blue","black","doms")
u.penGet()
```

```python
class User1:
    def __init__(self,pen,pencil):
        print("Constructor of User1")
        self.pen=pen
        self.pencil=pencil
    def penGet(self):
        print("pen is working :",self.pen)
    def pencilGet(self):
        print("pencil is working :",self.pencil)

class User2(User1):
    def __init__(self,pen,pencil,erasor):
        print("Constructer of User2")
       # self.pen=pen
       # self.pencil=pencil
        super().__init__(pen,pencil)
        self.erasor=erasor

#object
u=User2("blue","black","doms")
u.penGet()
```
