Handle missing value

```python

import numpy as np
import pandas as pd
data={
    "experience":[1,2,3,4,5,np.nan],
    "salary":[100,200,300,400,500,np.nan],
    "name":['hello','dheeraj','Nikhil','Raj','vinay','aditya']
}
df=pd.DataFrame(data)
print(df)
```

```python
#null value filter | find out
print(df.isnull())

#sum of null
print(df.isnull().sum())

#dropna | drop complete row
df.dropna()
print(df)

#fillna | nll -> fill
df.fillna(5)
```
